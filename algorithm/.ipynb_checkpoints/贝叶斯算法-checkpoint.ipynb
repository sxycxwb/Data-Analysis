## 贝叶斯算法

贝叶斯原理与其他统计学推断方法不同，它建立在主观判断的基础上：在我们不了解所有客观事实的情况下，同样可以先估计一个值，然后根据实际结果不断进行修正。

#### 贝叶斯原理的重要概念

先验概率：通过经验来判断事情发生的概率

后验概率：发生结果之后，推测原因的概率

条件概率：事件A在另外一个事件B已经发生的条件下的发生概率

似然函数：统计参数的函数。把概率模型的训练过程理解为求参数估计的过程

#### 贝叶斯原理

贝叶斯原理就是求解后验概率


#### 朴素贝叶斯

每个类别的概率和每个属性的条件概率

贝叶斯原理是最大概念，解决了概率论中的“逆向概率”问题，在此基础上设计了贝叶斯分类器，朴素贝叶斯是分类器中的一种，要求属性相互独立

#### 朴素贝叶斯流程

准备阶段：确定特征属性-->获取训练样本

训练阶段：计算每个类别P(cj)-->对每个特征属性计算所有划分的条件概率  【计算每个类别在训练样本中的出现频率及每个特征属性划分对每个类别的条件概率】

应用阶段：对每个类别计算P(Ai|Cj)P(Cj)-->P(Ai|Cj)P(Cj)最大值时的分类Cj即为类别

## sklearn下的三种朴素贝叶斯算法

高斯朴素贝叶斯（GaussianNB）：特征变量是连续变量，符合高斯分布，比如物体的长度，鞋码大小

多项式朴素贝叶斯（MultinomialNB）：特征变量是离散变量，符合多项分布，在文档分类中特征变量体现在一个单词出现的次数，或者是单词的 TF-IDF 值等

伯努利朴素贝叶斯（BernoulliNB）：特征变量是布尔变量，符合 0/1 分布，在文档分类中特征是单词是否出现
